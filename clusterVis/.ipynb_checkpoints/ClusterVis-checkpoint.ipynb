### Dados Coletados 1º bateria de teste.

## Com a normalização. 

#### Nessa tabelas temos os o número de grupos utilizados nesses experimento. Para isso, foram dez iteração para cada número k de groups.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

wi = pd.read_csv('withNormilzation.csv')
di = {}

for group in range(2,11):
    base_filtered = wi[wi['group'] == group]
    di[group] = base_filtered['dunn'].values
    
df = pd.DataFrame(di)

print(df)
```

```python
for i in range(2,11):
    x = [ i for j  in range(10)]
    y = df[i]
    plt.scatter(x,y)
plt.show()
```

#### Nesse gráficos temos as dez  iteração de cada número k distribuída de plano cartesiano.

## 1º. Índice de Dunn em função do número de Clusters.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def getXY(file):
    groups= []
    maximuns = []
    wi = pd.read_csv(file)
    for group in range(2,11):
        base_filtered = wi[wi['group'] == group]
        groups.append(group)
        maximuns.append(base_filtered['dunn'].max())
    return groups, maximuns


groups,maximuns = getXY('withNormilzation.csv')
plt.plot(groups, maximuns)

groups,maximuns = getXY('withoutNormilzation.csv')
plt.plot(groups,maximuns)
plt.xlabel("Número de Clusters")
plt.ylabel("Índice de Dunn")
```

<h3 style='color: orange'>*Sem a normalização</h3>
<h3 style='color: blue'>*Com a normalização</h3>

## 2º. Índice de Dunn em função do número de Clusters.

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def getXY(file):
    groups= []
    maximuns = []
    wi = pd.read_csv(file)
    for group in range(2,11):
        base_filtered = wi[wi['group'] == group]
        groups.append(group)
        maximuns.append(base_filtered['dunn'].max())
    return groups, maximuns

groups,maximuns = getXY('withNormilzation2.csv')
plt.plot(groups, maximuns)

groups,maximuns = getXY('withoutNormilzation2.csv')
plt.plot(groups,maximuns)
plt.xlabel("Número de Clusters")
plt.ylabel("Índice de Dunn")
plt.show()
```

<h3 style='color: orange'>*Sem a normalização</h3>
<h3 style='color: blue'>*Com a normalização</h3>
